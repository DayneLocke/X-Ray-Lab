```python
import pandas as pd


data = pd.read_csv('Lab4data.csv', usecols=['Angle', 'No Filter', 'Zr', 'Mo', 'Ag'])

#Zr,Mo,Ag
mass=[40,42,47]
kedge=[68.7,61.9,49.6]
kerror=[1.0,0.8,0.6]

angle=data['Angle'].tolist()
nofilter=data['No Filter'].tolist()
zr=data['Zr'].tolist()
mo=data['Mo'].tolist()
ag=data['Ag'].tolist()

transmittancezr=[]
transmittancemo=[]
transmittanceag=[]

for i in range(len(angle)):
  transmittancezr.append(zr[i]/nofilter[i])
  transmittancemo.append(mo[i]/nofilter[i])
  transmittanceag.append(ag[i]/nofilter[i])

print(transmittancezr)
print(transmittancemo)
print(transmittanceag)


print(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(angle,nofilter,label='No Filter',color='orange')
plt.plot(angle,zr,label='Zr',color='blue')
plt.plot(angle,mo,label='Mo',color='red')
plt.plot(angle,ag,label='Ag',color='green')
plt.axvline(x=6.6, color='black', linestyle='--')
plt.axvline(x=7.4, color='black', linestyle='--')
plt.xlabel(r'Angle($^\circ$)')
plt.ylabel(r'Counts ($s^{-1}$)')
plt.xlim(3.7,7.5)
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(angle,transmittancezr,label='Zr',color='blue')
plt.plot(angle,transmittancemo,label='Mo',color='red')
plt.plot(angle,transmittanceag,label='Ag',color='green')
plt.xlabel(r'Angle($^\circ$)')
plt.ylabel('Transmittance')
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.5)
plt.show()
```

```python
from scipy.stats import linregress

slope, intercept, r_value, p_value, std_err = linregress(kedge,mass)
masses=range(40,81)
line = [slope * x + intercept for x in masses]



plt.figure(figsize=(8, 6))
plt.scatter(kedge,mass,color='black')
plt.errorbar(kedge,mass,yerr=kerror,fmt='o',color='black')
plt.plot(masses, line, color='red', label=f'Best Fit Line\n$R^2$ = 0.924')
plt.ylim(35,50)
plt.xlim(40,80)
plt.xlabel('Atmoic Mass')
plt.ylabel('Kedge')
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.5)
plt.show()
```
